```python
import pandas as pd
import matplotlib.pyplot as plt
import torch
```

```python
df = pd.read_csv('data/student_performance.csv')
df.describe()
```

```python
df['weekly_self_study_hours'].value_counts().shape
```

```python
plt.figure(figsize=(50, 5))
df['weekly_self_study_hours'].value_counts().sort_index().plot(kind='bar')
plt.title('      ')
plt.xlabel(' ')
plt.ylabel(' ')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(30, 10))
df['class_participation'].value_counts().sort_index().plot(kind='bar')
plt.title('     ')
plt.xlabel('  ')
plt.ylabel(' ')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
